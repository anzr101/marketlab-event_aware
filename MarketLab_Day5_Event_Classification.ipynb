```python
# ============================================================================
# MARKETLAB DAY 5 - EVENT CLASSIFICATION SYSTEM
# ============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import json
from datetime import datetime, timedelta

print("🔥 MARKETLAB DAY 5 - EVENT CLASSIFICATION")
print("="*80)
print("🌍 Building Geopolitical Intelligence Layer")
print("="*80 + "\n")

# ============================================================================
# EVENT TAXONOMY
# ============================================================================

print("="*80)
print("PART 1: EVENT TAXONOMY & CLASSIFICATION")
print("="*80 + "\n")

# Define event categories and their market impact patterns
EVENT_TAXONOMY = {
    'geopolitical': {
        'description': 'Wars, conflicts, international tensions, trade disputes',
        'keywords': ['war', 'conflict', 'sanctions', 'trade war', 'tension', 'military',
                    'invasion', 'attack', 'crisis', 'dispute'],
        'typical_impact': 'high',
        'affected_sectors': ['defense', 'oil', 'commodities', 'safe_haven'],
        'volatility_increase': 'high'
    },
    'economic_policy': {
        'description': 'Central bank decisions, fiscal policy, interest rates',
        'keywords': ['interest rate', 'fed', 'rbi', 'monetary policy', 'inflation',
                    'rate hike', 'rate cut', 'quantitative easing', 'budget', 'gdp'],
        'typical_impact': 'high',
        'affected_sectors': ['banking', 'real_estate', 'all'],
        'volatility_increase': 'medium'
    },
    'regulatory': {
        'description': 'New regulations, policy changes, government actions',
        'keywords': ['regulation', 'law', 'ban', 'restriction', 'policy', 'government',
                    'tax', 'subsidy', 'approval', 'compliance'],
        'typical_impact': 'medium',
        'affected_sectors': ['sector_specific'],
        'volatility_increase': 'medium'
    },
    'corporate': {
        'description': 'Earnings, mergers, acquisitions, scandals',
        'keywords': ['earnings', 'merger', 'acquisition', 'ceo', 'bankruptcy', 'scandal',
                    'lawsuit', 'dividend', 'buyback', 'restructuring'],
        'typical_impact': 'medium',
        'affected_sectors': ['company_specific'],
        'volatility_increase': 'low'
    },
    'natural_disaster': {
        'description': 'Earthquakes, floods, pandemics, climate events',
        'keywords': ['earthquake', 'flood', 'pandemic', 'covid', 'hurricane', 'drought',
                    'climate', 'disaster', 'outbreak', 'emergency'],
        'typical_impact': 'medium',
        'affected_sectors': ['insurance', 'infrastructure', 'healthcare'],
        'volatility_increase': 'high'
    },
    'technological': {
        'description': 'Breakthroughs, disruptions, cyber events',
        'keywords': ['ai', 'technology', 'innovation', 'cyber attack', 'data breach',
                    'disruption', 'automation', 'breakthrough'],
        'typical_impact': 'low',
        'affected_sectors': ['tech', 'innovation'],
        'volatility_increase': 'low'
    }
}

print("📊 Event Taxonomy Created:")
for category, details in EVENT_TAXONOMY.items():
    print(f"\n   {category.upper()}:")
    print(f"   • Description: {details['description']}")
    print(f"   • Impact Level: {details['typical_impact']}")
    print(f"   • Volatility: {details['volatility_increase']}")

# Save taxonomy
with open(RESULTS_DAY5 / 'event_taxonomy.json', 'w') as f:
    json.dump(EVENT_TAXONOMY, f, indent=2)

print("\n✅ Taxonomy saved!")

# ============================================================================
# EVENT CLASSIFICATION FUNCTION
# ============================================================================

print("\n" + "="*80)
print("PART 2: EVENT CLASSIFIER")
print("="*80 + "\n")

def classify_event(event_text):
    """
    Classify an event into categories and assess impact

    Returns:
        - categories: List of matching categories
        - primary_category: Most relevant category
        - impact_score: 0-10 scale
        - confidence: Classification confidence
    """

    event_lower = event_text.lower()

    # Find matching categories
    category_scores = {}

    for category, details in EVENT_TAXONOMY.items():
        # Count keyword matches
        matches = sum(1 for keyword in details['keywords'] if keyword in event_lower)

        if matches > 0:
            # Weight by impact level
            impact_weight = {'high': 3, 'medium': 2, 'low': 1}[details['typical_impact']]
            category_scores[category] = matches * impact_weight

    if not category_scores:
        return {
            'categories': ['unknown'],
            'primary_category': 'unknown',
            'impact_score': 5,
            'confidence': 0.0,
            'sentiment': 'neutral'
        }

    # Sort by score
    sorted_categories = sorted(category_scores.items(), key=lambda x: x[1], reverse=True)

    primary_category = sorted_categories[0][0]
    primary_score = sorted_categories[0][1]

    # Calculate impact score (0-10)
    impact_map = {'high': 8, 'medium': 5, 'low': 3}
    base_impact = impact_map[EVENT_TAXONOMY[primary_category]['typical_impact']]

    # Adjust by number of keywords matched
    impact_score = min(10, base_impact + primary_score)

    # Confidence based on how clear the classification is
    total_score = sum(category_scores.values())
    confidence = primary_score / total_score if total_score > 0 else 0

    # Simple sentiment analysis
    negative_words = ['crisis', 'war', 'attack', 'crash', 'fall', 'decline', 'loss',
                     'ban', 'restriction', 'scandal', 'bankruptcy', 'disaster']
    positive_words = ['growth', 'rise', 'gain', 'profit', 'surge', 'breakthrough',
                     'approval', 'expansion', 'recovery', 'boom']

    neg_count = sum(1 for word in negative_words if word in event_lower)
    pos_count = sum(1 for word in positive_words if word in event_lower)

    if neg_count > pos_count:
        sentiment = 'negative'
    elif pos_count > neg_count:
        sentiment = 'positive'
    else:
        sentiment = 'neutral'

    return {
        'categories': [cat for cat, score in sorted_categories],
        'primary_category': primary_category,
        'impact_score': impact_score,
        'confidence': confidence,
        'sentiment': sentiment,
        'keyword_matches': primary_score
    }

# Test with examples
print("🧪 Testing Event Classifier:\n")

test_events = [
    "Russia-Ukraine war escalates, oil prices surge",
    "Fed raises interest rates by 0.75%, markets react",
    "New COVID variant detected, travel restrictions imposed",
    "Apple announces record earnings, stock jumps 5%",
    "Government announces new tax on tech companies",
    "AI breakthrough: New model surpasses human performance"
]

test_results = []

for i, event in enumerate(test_events, 1):
    result = classify_event(event)
    test_results.append({
        'event': event,
        **result
    })

    print(f"{i}. EVENT: {event}")
    print(f"   → Category: {result['primary_category']}")
    print(f"   → Impact: {result['impact_score']}/10")
    print(f"   → Sentiment: {result['sentiment']}")
    print(f"   → Confidence: {result['confidence']:.2f}")
    print()

# ============================================================================
# HISTORICAL EVENT DATABASE (2022-2024)
# ============================================================================

print("="*80)
print("PART 3: HISTORICAL EVENT DATABASE")
print("="*80 + "\n")

# Key events that affected markets during backtest period
HISTORICAL_EVENTS = [
    {
        'date': '2022-02-24',
        'event': 'Russia invades Ukraine, global oil shock',
        'category': 'geopolitical',
        'impact': 10,
        'affected_stocks': ['all'],
        'market_reaction': 'sharp decline'
    },
    {
        'date': '2022-03-16',
        'event': 'Fed raises rates 0.25%, first hike since 2018',
        'category': 'economic_policy',
        'impact': 8,
        'affected_stocks': ['banking', 'all'],
        'market_reaction': 'volatile'
    },
    {
        'date': '2022-06-15',
        'event': 'Fed raises rates 0.75%, largest hike since 1994',
        'category': 'economic_policy',
        'impact': 9,
        'affected_stocks': ['banking', 'real_estate', 'all'],
        'market_reaction': 'decline'
    },
    {
        'date': '2022-09-21',
        'event': 'Fed raises rates another 0.75%, inflation fight continues',
        'category': 'economic_policy',
        'impact': 8,
        'affected_stocks': ['banking', 'all'],
        'market_reaction': 'decline'
    },
    {
        'date': '2023-03-10',
        'event': 'Silicon Valley Bank collapses, banking crisis fears',
        'category': 'corporate',
        'impact': 9,
        'affected_stocks': ['banking', 'tech'],
        'market_reaction': 'panic selling'
    },
    {
        'date': '2023-05-03',
        'event': 'Fed raises rates to 5.0-5.25%, highest since 2007',
        'category': 'economic_policy',
        'impact': 7,
        'affected_stocks': ['banking', 'all'],
        'market_reaction': 'mixed'
    },
    {
        'date': '2023-11-14',
        'event': 'US inflation cools to 3.2%, rate hike pause expected',
        'category': 'economic_policy',
        'impact': 7,
        'affected_stocks': ['all'],
        'market_reaction': 'rally'
    },
    {
        'date': '2024-01-31',
        'event': 'Fed holds rates steady, signals potential cuts',
        'category': 'economic_policy',
        'impact': 6,
        'affected_stocks': ['all'],
        'market_reaction': 'rally'
    },
    {
        'date': '2024-07-31',
        'event': 'India budget 2024: Tax changes announced',
        'category': 'regulatory',
        'impact': 5,
        'affected_stocks': ['all'],
        'market_reaction': 'volatile'
    },
    {
        'date': '2024-09-18',
        'event': 'Fed cuts rates 0.50%, first cut since 2020',
        'category': 'economic_policy',
        'impact': 8,
        'affected_stocks': ['all'],
        'market_reaction': 'rally'
    }
]

# Create event database
events_df = pd.DataFrame(HISTORICAL_EVENTS)
events_df['date'] = pd.to_datetime(events_df['date'])

print(f"📅 Historical Event Database: {len(events_df)} major events\n")
print(events_df[['date', 'event', 'category', 'impact']].to_string(index=False))

# Save database
events_df.to_csv(RESULTS_DAY5 / 'historical_events.csv', index=False)
print(f"\n✅ Event database saved!")

# ============================================================================
# VISUALIZATION
# ============================================================================

print("\n" + "="*80)
print("PART 4: VISUALIZATIONS")
print("="*80 + "\n")

print("🎨 Creating visualizations...")

# Create timeline visualization
fig, axes = plt.subplots(3, 1, figsize=(16, 12))

# Plot 1: Event timeline
events_df_sorted = events_df.sort_values('date')
colors = {'geopolitical': 'red', 'economic_policy': 'blue', 'corporate': 'green',
          'regulatory': 'orange', 'natural_disaster': 'purple'}

for idx, row in events_df_sorted.iterrows():
    color = colors.get(row['category'], 'gray')
    axes[0].scatter(row['date'], row['impact'], s=200, alpha=0.6, color=color)
    axes[0].annotate(row['event'][:30] + '...',
                    (row['date'], row['impact']),
                    fontsize=8, rotation=45, ha='right')

axes[0].set_ylabel('Impact Score', fontsize=12)
axes[0].set_title('Major Events Timeline (2022-2024)', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(0, 11)

# Plot 2: Events by category
category_counts = events_df['category'].value_counts()
axes[1].bar(category_counts.index, category_counts.values, alpha=0.7)
axes[1].set_ylabel('Number of Events', fontsize=12)
axes[1].set_title('Events by Category', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3, axis='y')
plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=45, ha='right')

# Plot 3: Average impact by category
avg_impact = events_df.groupby('category')['impact'].mean().sort_values(ascending=False)
axes[2].barh(avg_impact.index, avg_impact.values, alpha=0.7, color='coral')
axes[2].set_xlabel('Average Impact Score', fontsize=12)
axes[2].set_title('Average Impact by Category', fontsize=14, fontweight='bold')
axes[2].grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.savefig(RESULTS_DAY5 / 'event_analysis.png', dpi=300, bbox_inches='tight')
plt.close()

print("✅ Timeline visualization created!")

# Test results visualization
test_df = pd.DataFrame(test_results)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Impact scores
axes[0].barh(range(len(test_df)), test_df['impact_score'], alpha=0.7)
axes[0].set_yticks(range(len(test_df)))
axes[0].set_yticklabels([e[:40] + '...' for e in test_df['event']], fontsize=9)
axes[0].set_xlabel('Impact Score', fontsize=12)
axes[0].set_title('Event Impact Scores', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3, axis='x')

# Category distribution
category_dist = test_df['primary_category'].value_counts()
axes[1].pie(category_dist.values, labels=category_dist.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Event Category Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig(RESULTS_DAY5 / 'classifier_test_results.png', dpi=300, bbox_inches='tight')
plt.close()

print("✅ Test results visualization created!")

# ============================================================================
# VALIDATION METRICS
# ============================================================================

print("\n" + "="*80)
print("PART 5: VALIDATION METRICS")
print("="*80 + "\n")

validation_metrics = {
    'event_database_size': len(events_df),
    'date_range': f"{events_df['date'].min()} to {events_df['date'].max()}",
    'categories': len(EVENT_TAXONOMY),
    'total_keywords': sum(len(cat['keywords']) for cat in EVENT_TAXONOMY.values()),
    'avg_impact_score': events_df['impact'].mean(),
    'high_impact_events': len(events_df[events_df['impact'] >= 8]),
    'test_events_classified': len(test_df),
    'avg_confidence': test_df['confidence'].mean()
}

print("📊 Validation Metrics:")
for metric, value in validation_metrics.items():
    print(f"   • {metric}: {value}")

# Save metrics
with open(RESULTS_DAY5 / 'validation_metrics.json', 'w') as f:
    json.dump(validation_metrics, f, indent=2)

print("\n✅ Validation metrics saved!")

# ============================================================================
# FINAL SUMMARY
# ============================================================================

print("\n" + "="*80)
print("🎉 DAY 5 COMPLETE!")
print("="*80)
print("✅ Event taxonomy created: 6 categories")
print("✅ Event classifier built and tested")
print("✅ Historical database: 10 major events (2022-2024)")
print("✅ Visualizations created: 2 plots")
print(f"✅ Files created: {len(list(RESULTS_DAY5.glob('*')))}")
print("="*80)
print(f"\n📂 Results saved in: /MarketLab_BEAST/results_day5/")
print("\n📊 Key files created:")
print("   • event_taxonomy.json")
print("   • historical_events.csv")
print("   • event_analysis.png")
print("   • classifier_test_results.png")
print("   • validation_metrics.json")
print("\n🎯 KEY CAPABILITIES:")
print("   • Classifies events into 6 categories")
print("   • Scores impact 0-10")
print("   • Analyzes sentiment (positive/negative/neutral)")
print(f"   • Average classification confidence: {test_df['confidence'].mean():.2f}")
print("\n✅ READY FOR DAY 6 - REAL-TIME INTELLIGENCE AGENT!")
```
